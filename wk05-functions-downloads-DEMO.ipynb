## Functions: Making Your Code Work Smarter

A **function** is a reusable block of code that performs a specific task. Instead of writing the same code over and over, you write it once in a function and call it whenever you need it.

Functions help you:
* **Stay DRY** (Don't Repeat Yourself) – write code once, use it many times
* **Organize your code** – break complex problems into smaller, manageable pieces
* **Make debugging easier** – fix issues in one place instead of hunting through repeated code
* **Build modular programs** – combine simple functions to accomplish complex tasks

A function runs only when something "invokes" or "calls" it, giving you control over when and how your code executes.

## ```Docstrings```

As our functions get more complicated, it gets harder to remember what the parameters are and what they actually do. ```Docstrings``` help us and our colleagues understand what a function does; what parameters you need; and what order they need to be in.

Info to include in a ```docstring```:

* A brief summary of what the function does.
* A description of each ```parameter``` and their expected data types.
* The `Returns` section specifies what value (if any) the function
returns.

````python
## shorten our previous function
## you might forget what arguments are needed for the function to work.
## you can add notes that appear on shift-tab as you call the function.

def pctChg(old_value, new_value):
    '''
    This function takes two numbers and returns the percent change
    para1 = is the older number
    para2 = is the newer number
    '''
    return (new_value - old_value)/ old_value * 100
````

### To use or not use ```functions```?


**Use a function when you:**
* **Repeat the same code** multiple times in your program
* **Need to perform a specific task** that can be named and isolated (like calculating a total or formatting data)
* **Want to break down a complex problem** into smaller, manageable pieces


**Don't use a function when:**
* **The code only runs once** and is simple enough to understand in place
* **Creating a function would make the code harder to read** (over-engineering)
* **The code is too tightly coupled** to its surrounding context to be meaningfully separated

#### Rule of thumb: 

**If you're about to copy-paste code, write a function instead!**

### Your Challenge

Write a function that makes a request to a website and returns the content as soup.

````python
## write code here        

## dependencies
import requests
from bs4 import BeautifulSoup

## function
def makeSoup(url):
    '''
    Makes soup of any html site
    para1 = url of site to be scraped
    Returns soup
    '''
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}
    response = requests.get(url, headers = headers)
    if response.status_code == 200:
        return BeautifulSoup(response.text, "html.parser")
    else:
        print(f"Your request returned {response.status_code}")

````

````python
## place url into variable
url = "https://sandeepmj.github.io/scrape-example-page/demo-text.html"
````

## ```Flatten``` our ```nested``` list

we can ```itertools``` to flatten a list of list.

```flattened_list = list(chain.from_iterable(list_of_lists))```

````python
## import itertools
from itertools import chain
````

````python
## flatten using itertools
flat_targets = list(chain.from_iterable(atags))
flat_targets
````
