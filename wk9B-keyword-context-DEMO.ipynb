```python
import regex as re
import pandas as pd
from rapidfuzz import fuzz
from functools import lru_cache
import spacy
```

```python
keywords = ['opioid',
 'prescription',
 'sexual',
 'abuse',
 'assault',
            "misconduct",
 'unlawful',
 'fraud',
 'controlled substances']
```

```python
target_pdfs
```

## Now iterate through all files

```python
## iterate through files
df_list = []
for i, target_pdf in enumerate(target_pdfs, start = 0):
    print(f"Processing {i} of {len(target_pdfs)}... ")
    text = pdf2image_ocr_text(target_pdf)
    print(f"Looking for context words... ")
    df_list.append(keyword_context_finder(
    content=text,
    terms=keywords,
    file_name= target_pdf,
    words_around=10))
    


    
    
```

```python
## how long is df_list?
len(df_list)
```

```python
## concat them
df = pd.concat(df_list, ignore_index= True)
```

```python
## get info
df.sample(20)
```
